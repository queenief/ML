## Normalizing the dataframe

```python
#Normalize the strings in data frame ,converting strings to lower case and replace blank and ?? with nan


from pandas.api.types import is_string_dtype, is_object_dtype

def df_normalize_strings(data):
    for col in data.columns:
        if is_string_dtype(data[col]) or is_object_dtype(data[col]):
            data[col] = data[col].str.lower()
            data[col] = data[col].fillna(np.nan)
            data[col] = data[col].replace('', np.nan)
            data[col]=  data[col].replace('??',np.nan)
```

```python
def sniff(data):
  with pd.option_context("display.max_colwidth",20):
    info=pd.DataFrame()
    info['data type']=data.dtypes
    info['precent missing']=data.isnull().sum()*100/len(data)
    info['No of Unique']=data.apply(lambda x:len(x.unique()))
    #info['Unique Values']=data.apply(lambda x: (x.unique()))
    return info.sort_values('data type')
```

```python
df=df.rename(columns={"Car/Truck Category - Cash for Clunkers Bill.":"Car_Truck_category"})
df=df.rename(columns={"$ You Spend over 5 years (increased amount spent in fuel costs over 5 years - on label) ":"Amount_spent_over_5yrs"})
df=df.rename(columns={"Var Valve Timing":"Var_Valve_Timing"})
df=df.rename(columns={"Max Ethanol % - Gasoline":"Max_Ethanol_gasoline"})
df=df.rename(columns={"Gas Guzzler Exempt (Where Truck = 1975 NHTSA truck definition)":"Gas_Guzzler_Exempt"})
df=df.rename(columns={"$ You Save over 5 years (amount saved in fuel costs over 5 years - on label) ":"Amount_saved_over_5_yrs"})
```

```python
# Functions to convert String to categorical and encoding the categorical variables with cat codes.
from pandas.api.types import is_categorical_dtype, is_string_dtype

def string_to_cat(data):
    for col in data.columns:
        if is_string_dtype(data[col]):
            df[col] = df[col].astype('category').cat.as_ordered()

def cat_to_catcode(data):
    for col in data.columns:
        if is_categorical_dtype(data[col]):
            df[col] = df[col].cat.codes + 1
```

```python
#Convering String to categorical features and encoding them with cat codes thus filling the missing values
string_to_cat(df)
cat_to_catcode(df)

df.isnull().sum().sort_values(ascending=False)
```

```python
def fix_missing_num(df, colname):
    df[colname+'_na'] = pd.isnull(df[colname])
    df[colname].fillna(df[colname].median(), inplace=True)
```

```python
fix_missing_num(df, 'Amount_saved_over_5_yrs')
fix_missing_num(df, 'Amount_spent_over_5yrs')
fix_missing_num(df, '4Dr Lugg Vol')
fix_missing_num(df, '4Dr Pass Vol')
fix_missing_num(df, 'Max_Ethanol_gasoline')
```
